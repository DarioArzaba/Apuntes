**Fuerza Electrica y Gravitacional**

* Calcular la razón entre la fuerza electrica y la fuerza gravitacional de dos protones.

\begin{gathered}
    F_e = \frac{kq^2}{r^2} \quad F_g = \frac{Gm^2}{r^2}\\
    \frac{F_e}{F_g} = \frac{kq^2}{Gm^2} \implies \frac{(9\times10^{9})(3.2\times10^{19})^2}{(6.67\times10^{-11})(6.64\times10^{-27})} = 3.1\times10^{35}
\end{gathered}

**Magnitud Fuerza en Linea**

\begin{gathered}
    q1 = 25nC \quad q2 = -75nC \quad r = 3cm\\
    F_{1->2}  = \frac{(9\times10^{-9})|(25\times10^{-9})(-75\times10^{-9})|}{(0.03 m)^2} = 0.019N(-\vec{i}) = - 0.019N\vec{i}\\
    F_{2->1}  = 0.019N(\vec{i}) = + 0.019N\vec{i}
\end{gathered}

**Suma Fuerzas en Linea**

\begin{gathered}
    q1 = 1nC \quad q2 = -3nC \quad q3 = 5nC \quad r_{13} = 2cm \quad r_{12} = 4cm\\
    F_{1->3}  = \frac{(9\times10^{-9})|(1\times10^{-9})(5\times10^{-9})|}{(0.02 m)^2} = 112\mu N \implies \vec{F_{1->3}}  = - 112 \mu N (\vec{i})\\
    F_{2->3}  = \frac{(9\times10^{-9})|(-3\times10^{-9})(5\times10^{-9})|}{(0.04 m)^2} = 84\mu N \implies \vec{F_{2->1}}  = + 84 \mu N (\vec{i})\\
    \vec{F_{12->3}} = - 112 \mu N (\vec{i}) + 84 \mu N (\vec{i}) = - 28 \mu N (\vec{i})
\end{gathered}

**Fuerzas superimpuestas**

\begin{gathered}

    F_{1->3}y + F_{2->3}y = 0 \land F_{1->3} = F_{2->3}\\

**Ecuacion campo Recta**

\begin{gathered}
    dE = \frac{k(dQ)}{r^2}\\
    r = \sqrt{x^2+y^2} \quad E = Ex \quad \lambda_{dy} = \frac{dQ}{dy}\\
    dy\lambda = dQ \implies dQ = \frac{Qdy}{2a} \implies dE = \frac{kQdy}{2ar^2}\\
    dE_x = (dE)(\cos(\theta)) \land \cos(\theta) = \frac{x}{R} \implies dE_x = \frac{kQdyx}{2ar^2R}\\
    E_x = \frac{kQ}{2a} \int \frac{xdy}{r^3} = \frac{kQ}{2a} \int_{-a}^{a} \frac{xdy}{(x^2+y^2)^{3/2}} = \int \frac{xdy}{(x^2)(1+\frac{y^2}{x^2})^{3/2}}\\
    y=x\tan(\theta) \land y^2 = x^2\tan^2(\theta) \land dy = x\sec^2(\theta)d\theta\\
    \int \frac{x(xsex^2(\theta))d\theta}{((x^2)(1+\frac{x^2\tan(\theta)}{x^2}))^{3/2}} \implies \int \frac{x^2sex^2(\theta)d\theta}{(x^3)(\sec^3(\theta))} \implies \int \frac{d\theta}{(x)(\sec(\theta))}\\
    \frac{1}{x} \int \frac{d\theta}{\sec(\theta)} \implies \frac{1}{x} \int \cos(\theta)d\theta \implies \frac{1}{x}\sin(\theta)\implies \frac{y}{xR}\\
    \frac{y}{x\sqrt{x^2+y^2}} \implies \frac{a - (-a)}{\sqrt{x^2+a^2}} \implies \frac{kQ2a}{2ax\sqrt{x^2+a^2}}\\
    E_x = \frac{kQ}{x\sqrt{x^2+a^2}} \land a = \frac{L}{2}\\
\end{gathered}

**Ecuacion campo Anillo**

\begin{gathered}
    dE = \frac{k(dQ)}{r^2}\\
    dE_x = \frac{kxdQ}{r^3} \implies E_x = \frac{kx}{r^3} \int dQ\\
    E_x = \frac{kxQ}{r^3} \land r = \sqrt{x^2+a^2}
\end{gathered}

**Ecuacion campo Disco**

\begin{gathered}
    dE = \frac{k(dQ)}{r^2}\\
    r + dr = dQ \land dA = 2\pi r dr\land \sigma = \frac{dQ}{dA} \implies dQ = \sigma dA = \sigma2\pi r dr\\
    E_x = \frac{kxQ}{r^3} \implies E_x = \int_{0}^{R} \frac{k2\pi\sigma rx dr}{r^3}\\
    E_x = \frac{\sigma x}{2\epsilon_0}\left[ 1 - \frac{1}{\sqrt{\frac{R^2}{x^2}+1}} \right] 
\end{gathered}

```python
q1 = widgets.FloatText(value=1, description='q1', disabled = False)
q2 = widgets.FloatText(value=1, description='q2', disabled = False)
uq1 = widgets.Dropdown(options=["C", "mC", "uC", "nC"], value="C", description="q1 Units: ")
uq2 = widgets.Dropdown(options=["C", "mC", "uC", "nC"], value="C", description="q2 Units: ")
ur = widgets.Dropdown(options=["m", "cm", "mm"], value="m", description="r Units: ")
x1 = widgets.FloatText(value=0, description='x1')
y1 = widgets.FloatText(value=0, description='y1')
x2 = widgets.FloatText(value=1, description='x2')
y2 = widgets.FloatText(value=0, description='y2')
campo = widgets.HTML(value="(Para calcular campo E usa q2 = 1C)")

def f(q1, q2, uq1, uq2, ur, x1, y1, x2, y2):
    
    R = math.hypot(y2-y1, x2-x1)
    
    # Cambiar unidades datos
    if (ur == "m"):
        R = R
    elif (ur == "cm"):
        R = R/100.0
    elif (ur == "mm"):
        R = R/1000.0
    if (uq1 == "C"):
        q1 = q1
    elif (uq1 == "mC"):
        q1 = q1/1000.0
    elif (uq1 == "uC"):
        q1 = q1/1000000.0
    elif (uq1 == "nC"):
        q1 = q1/1000000000.0
    if (uq2 == "C"):
        q2 = q2
    elif (uq2 == "mC"):
        q2 = q2/1000.0
    elif (uq2 == "uC"):
        q2 = q2/1000000.0
    elif (uq2 == "nC"):
        q2 = q2/1000000000.0
        
    f = Decimal(((8.98755178e9)*(q1)*(q2))/(abs(R**2)))

    # r12 vector que va de q1 a q2
    r12i = ((x2)-(x1))
    r12j = ((y2)-(y1))
    # r distancia que va de q1 a q2
    r = math.sqrt( (r12i)**2 + (r12j)**2 )
    # a12 vector unitario que va de q1 a q2
    a12i = r12i/r
    a12j = (r12j/r)
    # Vector fuerza en q2 generada por q1
    f12i = Decimal(f)*Decimal(a12i)
    f12j = Decimal(f)*Decimal(a12j)

    # r21 vector que va de q2 a q1
    r21i = ((x1)-(x2))
    r21j = ((y1)-(y2))
    # r distancia que va de q1 a q2
    r = math.sqrt( (r21i)**2 + (r21j)**2 )
    # a12 vector unitario que va de q1 a q2
    a21i = r21i/r
    a21j = (r21j/r)
    # Vector fuerza en q2 generada por q1
    f21i = Decimal(f)*Decimal(a21i)
    f21j = Decimal(f)*Decimal(a21j)

    ang12 = np.degrees(np.arctan2(*[float(f12j), float(f12i)])) % 360.0
    ang21 = np.degrees(np.arctan2(*[float(f21j), float(f21i)])) % 360.0
    display(Markdown(r'$F = \frac{{(9\times10^9 Nm^2/C^2)({} C)({} C)}}{{({} m)^2}} = {:.4E} N $'.format(q1, q2, R, abs(f))))
    display(Markdown(r'$\vec{{F_{{q1->q2}}}} = ({:.3g} N)({:.3g}\hat{{i}}+({:.3g}\hat{{j}})) = (({:.3g})\hat{{i}}({:.3g})\hat{{j}}) [N] $'.format(f, a12i, a12j, f12i, f12j)))
    display(Markdown(r'$\theta = \tan^{{-1}}(\frac{{{:.3g}}}{{{:.3g}}}) = {:.3g}°$'.format(f12i, f12j, ang12)))
    display(Markdown(r'$\vec{{F_{{q2->q1}}}} = ({:.3g} N)({:.3g}\hat{{i}}+({:.3g}\hat{{j}})) = (({:.3g})\hat{{i}}({:.3g})\hat{{j}}) [N] $'.format(f, a21i, a21j, f21i, f21j)))
    display(Markdown(r'$\theta = \tan^{{-1}}(\frac{{{:.3g}}}{{{:.3g}}}) = {:.3g}°$'.format(f21i, f21j, ang21)))

out = widgets.interactive_output(f, {'q1': q1, 'q2': q2, 'uq1': uq1, 'uq2': uq2,
                                     'ur': ur, 'x1': x1, 'y1': y1, 'x2': x2, 'y2': y2 })

tab = widgets.Tab()
datos = widgets.HBox([widgets.VBox([campo, q1, q2, uq1, uq2, ur]), out])
contenidoUnidades = widgets.HBox([x1, y1, x2, y2])
contenidoUnidades.layout=Layout(width='100%',display='inline-flex',flex_flow='row wrap')
unidades = widgets.VBox([contenidoUnidades])
tab.children = [datos, unidades]
tab.set_title(index = 0, title = "Datos")
tab.set_title(index = 1, title = "Unidades")
tab
```
